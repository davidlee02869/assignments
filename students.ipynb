```python
def students():
    
    student_last_name = input("Student's Last Name")

    if student_last_name == "ZZZ":
        
        print("quit")
        
        return
        
    else:
        
        student_first_name = input("Student's First Name")
        
        student_GPA = float(input("Student's GPA"))

        if student_GPA >= 3.5:
            
            print(f"{student_first_name} {student_last_name} has made it on the Dean's List")
            
        elif student_GPA >= 3.25:
            
            print(f"{student_first_name} {student_last_name} has made it on the Honor Roll")
            
        else:
            
            print(f"{student_first_name} {student_last_name} has not made it on the Dean's List or Honor Roll")


while True:
    
    students()
```
